```python
import pandas as pd
```

```python
estabelecimentos = pd.read_csv('data/estabelecimentos.csv', index_col=None)
pracas = pd.read_csv('data/praças.csv', index_col=None)
pontos_onibus = pd.read_csv('data/pontos_onibus.csv', index_col=None)
pontos_onibus_cobertos = pd.read_csv('data/pontos_cobertos.csv', index_col=None)
escolas = pd.read_csv('data/escolas.csv', index_col=None)
crimres = pd.read_csv('data/crimes.csv', index_col=None)
```

## Cleaning estabelecimentos

```python
estabelecimentos.head()
```

```python
estabelecimentos = estabelecimentos.drop(columns=['Num_inscrição', 'Nome'])
```

```python
estabelecimentos
```

## Cleaning praças

```python
pracas
```

```python
# dealing with names in different rows

list_pracas = []

for row in pracas.index:
    text = pracas.iloc[row][0].split(' ')
    
    if text[0].isupper():
        list_pracas.append(pracas.iloc[row][0])
    
    elif (len(text) > 1) and (text[1].isupper()) and (len(text[1]) > 3):
        list_pracas.append(pracas.iloc[row][0])
        
    else:
        list_pracas[-1] += pracas.iloc[row][0]
        
```

```python
pracas_names = [text.split('-', 1)[0] for text in list_pracas]
```

```python
pracas_description = [text.split('-', 1)[1] for text in list_pracas]
```

```python
pracas = pd.DataFrame({'Nome': pracas_names, 'Descriçao':pracas_description})
```
