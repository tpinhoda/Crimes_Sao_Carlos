## Plotting pontos_onibus
#### Todo
    - Compare locations: pontos_onibus and pontos_onibus_cobertos

```python
import folium

map = folium.Map(location=[-22.010147, -47.890706], zoom_start=13)
```

```python
pontos_onibus
```

```python
latitudes = list(pontos_onibus['Latitude'])
longitudes = list(pontos_onibus['Longitude'])

for lat, lng in zip(latitudes, longitudes):
    folium.Marker(location=[lat, lng]).add_to(map)
```
