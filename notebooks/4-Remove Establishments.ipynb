```python
import pandas as pd
```

```python
estabelecimentos = pd.read_csv('data/cleaned_features/estabelecimentos.csv', index_col=None)
```

```python
estabelecimentos.columns
```

```python
estabelecimentos['Tipo'].unique()
```

```python
bares = estabelecimentos[estabelecimentos['Tipo'].str.find('BAR') != -1]
bares = bares.drop(columns=['Tipo'])
bares
```

```python
restaurantes = estabelecimentos[estabelecimentos['Tipo'].str.find('RESTAURANTE') != -1]
restaurantes = restaurantes.drop(columns=['Tipo'])
restaurantes
```

```python
bares_restaurantes = pd.concat([bares.assign(Tipo='Bar'), restaurantes.assign(Tipo='Restaurante')])
bares_restaurantes = bares_restaurantes.reset_index()
bares_restaurantes
```

## Converting to csv

```python
bares.to_csv(r'data/cleaned_features/bares.csv', index=False)
```

```python
restaurantes.to_csv(r'data/cleaned_features/restaurantes.csv', index=False)
```

```python
bares_restaurantes.to_csv(r'data/cleaned_features/bares_restaurantes.csv', index=False)
```
