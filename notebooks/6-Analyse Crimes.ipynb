```python
crimes = pd.read_csv('data/crimes_with_no_geo_reference.csv', index_col=None)
```

```python
crimes
```

```python
map = folium.Map(location=[-22.010147, -47.890706], zoom_start=13)
```

```python
def in_San_Charles(lat, lng):
    return lat < -21 and lat > -23 and lng > -48 and lng < -47 
```

```python
latitudes = list(crimes['LATITUDE'])
longitudes = list(crimes['LONGITUDE'])
labels = list(crimes['Num_BO'])
out_of_bounds = 0

for lat, lng, label in zip(latitudes, longitudes, labels):
    if not in_San_Charles(lat, lng):
        out_of_bounds += 1
        folium.Marker(location=[lat, lng], tooltip=label, popup=[lat, lng]).add_to(map)
        
print(out_of_bounds)
```
